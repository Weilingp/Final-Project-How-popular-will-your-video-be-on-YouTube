```python
import pandas as pd
import os
import nltk
import re
import seaborn as sns
import wordninja
from transformers import pipeline
#from transformers import AutoModelFOrSequenceClassification
#from nltk import SentimentIntensityAnalyser
from tqdm.notebook import tqdm
from scipy.special import softmax
from textblob import TextBlob
from wordcloud import wordcloud
import numpy as np
```

```python
GBcomments = pd.read_csv("created_csvfiles/GBcomments.csv", error_bad_lines=False) #
```

```python
GBcomments.head()
```

```python
GBcomments.shape
```

```python
#cleaning the comments
```

```python
GBcomments["clean_comment"] = GBcomments["comment_text"].apply(lambda s: re.sub(r'[^A-Za-z0-9 ]+', ' ', str(s)))
#GBcomments.head()
```

```python
GBcomments["clean_comment"] = GBcomments["clean_comment"].apply(lambda s: ' '.join(wordninja.split(s)))
GBcomments.head()
```

```python
#polarity for comment analysm and lowercase
```

```python
def lower_comment(comment):
    return comment.lower()
```

```python
GBcomments["clean_comment_lower"] = GBcomments["clean_comment"].apply(lower_comment)
```

```python
def polarityfunc(text):
    return TextBlob(text).sentiment.polarity
```

```python
GBcomments["polarity"] = GBcomments["clean_comment_lower"].apply(polarityfunc)
GBcomments.set_index('video_id')
```

```python
#comment analysm labeling
```

```python
def comment_anlysm(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
GBcomments['comment_anlysm']= GBcomments['polarity'].apply(comment_anlysm)
GBcomments
```

```python
#grouping by comment sentiments: pos, neg, neu
```

```python
df_positiv_comments = GBcomments.loc[GBcomments["polarity"] > 0]
df1 = df_positiv_comments.groupby(["video_id"])["comment_anlysm"].size().reset_index(name ="positive_comments")
```

```python
df_negative_comments = GBcomments.loc[GBcomments["polarity"] < 0]
df2 = df_negative_comments.groupby(["video_id"])["comment_anlysm"].size().reset_index(name = "negative_comments")
```

```python
df_neutral_comments = GBcomments.loc[GBcomments["polarity"] == 0]
df3 = df_neutral_comments.groupby(["video_id"])["comment_anlysm"].size().reset_index(name = "Neutral_comments")
```

```python
#merging the counts of pos neg and neu commments per video
```

```python
from functools import reduce
dfs = [df1, df2, df3]

#merge all DataFrames into one
final_df = reduce(lambda  left,right: pd.merge(left,right,on=['video_id'],
                                            how='outer'), dfs)
final_df
```

```python
final_df.fillna(0)
```

```python
#fun to plot comments analysis per video
```

```python
def ploting_viewer_sent_per_video(videoid):
    
    video = final_df.loc[final_df["video_id"] == videoid, :]
    video.plot.bar()
    
    return video
```
