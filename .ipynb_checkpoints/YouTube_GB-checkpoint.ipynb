# sentiment analysis on : Titles

```python
# title_senti_analysis = {}

# for i,row in tqdm(df.iterrows(),total=len(df.title)):
#     text = row['title']
#     myid = row['video_id']
#     result[myid]=sia.polarity_scores(df)
    
# title_senti_analysis

result = {}
for i,row in tqdm(df.iterrows(),total=len(df)):
    text = row['title']
    video_id = row['video_id']
    result[video_id]=sia.polarity_scores(text)
    
result = pd.DataFrame(result)
result = result.T
score = result.compound

def getAnalysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
result['Analysis_title']= result['compound'].apply(getAnalysis)
```

```python
result
```

```python
result.reset_index(inplace=True)
result.rename(columns = {'index':'video_id',
                         'neg':'title_negative',
                         'neg':'title_negative',
                         'neu':'title_neutral',
                         'pos':'title_positive',
                         'compound':'title_compound'}, inplace = True)
df = pd.merge(df,result)
df[['video_id','title_negative', 'title_neutral', 'title_positive', 'title_compound','Analysis_title']].head(5)
```

# sentiment analysis on: Tags

```python
result_tags = {}
for i,row in tqdm(df.iterrows(),total=len(df)):
    text = row['tags']
    video_id = row['video_id']
    result_tags[video_id]=sia.polarity_scores(text)
    
result_tags = pd.DataFrame(result_tags)
result_tags = result_tags.T
score_tags = result_tags['compound']

```

```python
def getAnalysis(score_tags):
    if score_tags < 0:
        return "Negative"
    elif score_tags == 0:
        return "Neutral"
    else:
        return "Positive"
result_tags['Analysis_tags']= result_tags['compound'].apply(getAnalysis)
result_tags = pd.DataFrame(result_tags)

result_tags
```

```python

result_tags.reset_index(inplace=True)
result_tags.rename(columns = {'index':'video_id',
                              'neg':'tags_negative',
                              'neu':'tags_neutral',
                              'pos':'tags_positive',
                              'compound':'tags_compound'}, inplace = True)
result_tags
```

```python
df = pd.merge(df,result_tags)
df[['video_id','tags_negative', 'tags_neutral', 'tags_positive', 'tags_compound','Analysis_tags']].head()
```
